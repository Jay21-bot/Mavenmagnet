## Data transformation

### Subtask:
Transform the `df_input` DataFrame to aggregate the data by 'Age' and 'Keyword' to match the structure of `df_output`.


**Reasoning**:
Aggregate the data in `df_input` by 'Age' and 'Keyword' to count the occurrences of each 'Keyword' for each 'Age' group, which will create the structure similar to `df_output`.



```python
df_transformed = df_input.pivot_table(index='Age', columns='Keyword', aggfunc='size', fill_value=0)
df_transformed = df_transformed.reset_index()
df_transformed = df_transformed.rename(columns={'Total': 'Grand Total'})
display(df_transformed)
```

## Calculate percentages

### Subtask:
Calculate the percentage columns in the transformed DataFrame to match the columns like 'Intender.1', 'Owner.1', and 'Rejector.1' in `df_output`.


**Reasoning**:
Calculate the percentage columns based on the 'Grand Total' for each age group.



```python
df_transformed['Intender.1'] = df_transformed['Intender'] / df_transformed['Grand Total']
df_transformed['Owner.1'] = df_transformed['Owner'] / df_transformed['Grand Total']
df_transformed['Rejector.1'] = df_transformed['Rejector'] / df_transformed['Grand Total']
df_transformed['Grand Total.1'] = df_transformed['Grand Total'] / df_transformed['Grand Total'].sum()

display(df_transformed)
```

## Handle missing values

### Subtask:
Address any missing values that might have been introduced during the transformation process.


**Reasoning**:
Inspect the transformed dataframe for missing values and handle them by filling NaN values in 'Age' column with a placeholder and filling NaN values in numeric columns with 0.



```python
print("Missing values before handling:")
print(df_transformed.isnull().sum())

df_transformed['Age'] = df_transformed['Age'].fillna('Unknown')
for col in ['Intender', 'Owner', 'Rejector', 'Grand Total', 'Intender.1', 'Owner.1', 'Rejector.1', 'Grand Total.1']:
    df_transformed[col] = df_transformed[col].fillna(0)

print("\nMissing values after handling:")
print(df_transformed.isnull().sum())
```

## Format the output

### Subtask:
Ensure the final DataFrame has the correct column names and order as `df_output`.


**Reasoning**:
Reorder the columns of df_transformed to match the order of columns in df_output and then display the head of the reordered dataframe.



```python
df_transformed = df_transformed.reindex(columns=df_output.columns)
display(df_transformed.head())
```

## Save to csv

### Subtask:
Save the transformed DataFrame to a CSV file.


**Reasoning**:
Save the transformed DataFrame to a CSV file named `df_output.csv`, excluding the index.



```python
df_transformed.to_csv('df_output.csv', index=False)
```
